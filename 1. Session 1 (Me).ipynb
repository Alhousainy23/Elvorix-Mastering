```python
# Check values for skewed 
def plot_hist_box(column_data,column_name=None):
    plt.style.use("ggplot")
    if column_name is None : column_name = column_data.name
    plt.figure(figsize=(12, 4))

    #Histogram
    
    plt.subplot(1, 2, 1)
    sns.histplot(column_data,kde=True,bins=20,color='skyblue')
    plt.title(f"Histogram of {column_name}")
    plt.xlabel(column_name)
    plt.ylabel("Frequency")
    
    #Boxplot
    plt.subplot(1, 2, 2)
    sns.boxplot(x= column_data,color='lightgreen')
    plt.title(f"Boxplot of {column_name}")
    plt.xlabel(column_name)
    
    plt.tight_layout()
    plt.show()
print("The Column Name when not add any name to column , return the same name ==> ",dataframe['mileage'].name)
import warnings
warnings.filterwarnings("ignore")
#Call The Function for the Milage Column
plot_hist_box(dataframe['mileage'])
```

### ⛽ Fuel Column (Categorical)

The **`fuel`** column is **categorical**, so we can fill the **NaN values** with the **mode** (the most frequent value). 📊✅

```python
# Used mode because fuel column is categorical data
data['fuel'] = data['fuel'].fillna(data['fuel'].mode()[0])
data['fuel'].isnull().sum()
```
